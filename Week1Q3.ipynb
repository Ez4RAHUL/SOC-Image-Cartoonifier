```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DataStock.csv')
```

```python
usefuldata = pd.DataFrame(df[['Date','Close']])
```

```python
usefuldata
```

```python
npClose = usefuldata['Close'].values
npDate = usefuldata['Date'].values
```

```python
npClose
```

```python
npDate
```

```python
n = npClose.size
k = 0
p = 0
avgmonth = []

for i in range(n-1):
    if npDate[i][0:2] == npDate[i+1][0:2]:
        k += 1
    else:
        avgmonth.append(np.average(npClose[p:p+k+1]))  # Include the current day's closing price
        p = i + 1
        k = 0

# Calculate average for the last month
avgmonth.append(np.average(npClose[p:]))

for i in avgmonth:
    print(i)
```
